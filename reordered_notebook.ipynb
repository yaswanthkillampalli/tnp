## 19. Python program for stack using list

```python
# Stack implementation using list
stack = []

# Push elements
stack.append(10)
stack.append(20)
stack.append(30)
stack.append(40)
print(f"Stack after pushing: {stack}")

# Pop element
popped = stack.pop()
print(f"Popped element: {popped}")
print(f"Stack after popping: {stack}")

# Peek element
if stack:
    print(f"Top element (peek): {stack[-1]}")

# Check if stack is empty
print(f"Is stack empty? {len(stack) == 0}")
```

## 20. Python program for queue using list

```python
# Queue implementation using list
queue = []

# Enqueue elements
queue.append(10)
queue.append(20)
queue.append(30)
queue.append(40)
print(f"Queue after enqueuing: {queue}")

# Dequeue element
dequeued = queue.pop(0)
print(f"Dequeued element: {dequeued}")
print(f"Queue after dequeuing: {queue}")

# Check front element
if queue:
    print(f"Front element: {queue[0]}")

# Check if queue is empty
print(f"Is queue empty? {len(queue) == 0}")
```

## 23. Access each character of string in forward and backward direction using while loop

```python
# Access characters in forward and backward direction using while loop
string = input("Enter a string: ")

print("Forward direction:")
i = 0
while i < len(string):
    print(string[i], end=" ")
    i += 1
print()

print("Backward direction:")
i = len(string) - 1
while i >= 0:
    print(string[i], end=" ")
    i -= 1
print()
```

## 28. Program for requirement: input - a-4 b-3 c-2, output - aaaa bbb cc

```python
# Convert input format a-4 b-3 c-2 to output aaaa bbb cc
input_str = input("Enter input (format: a-4 b-3 c-2): ")
pairs = input_str.split()
result = []

for pair in pairs:
    char, count = pair.split('-')
    count = int(count)
    result.append(char * count)

output = ' '.join(result)
print(f"Input: {input_str}")
print(f"Output: {output}")
```

## 29. Program for requirement: input - a-4 k-3 h-2, output - ae kn bd

```python
# Convert input a-4 k-3 h-2 to output ae kn bd (character + offset positions)
input_str = input("Enter input (format: a-4 k-3 h-2): ")
pairs = input_str.split()
result = []

for pair in pairs:
    char, count = pair.split('-')
    count = int(count)
    # Get character and the character at count positions ahead
    output_pair = char + chr(ord(char) + count)
    result.append(output_pair)

output = ' '.join(result)
print(f"Input: {input_str}")
print(f"Output: {output}")
```

## 40. Process student marks based on who passed (pass percentage is 35) add 5 grace marks to all students and find total marks of all passed students

```python
# Process student marks: add grace marks, determine pass/fail, find total of passed students
n = int(input("Enter number of students: "))
students = {}
grace_marks = 5
pass_percentage = 35

# Input student marks
for i in range(n):
    name = input(f"Enter name of student {i+1}: ")
    marks = int(input(f"Enter marks for {name}: "))
    students[name] = marks

# Add grace marks and determine pass/fail
print("\nStudent Results:")
total_passed = 0
for name, marks in students.items():
    marks_with_grace = marks + grace_marks
    status = "PASSED" if marks_with_grace >= pass_percentage else "FAILED"
    print(f"{name}: {marks} + {grace_marks} (grace) = {marks_with_grace} - {status}")
    if marks_with_grace >= pass_percentage:
        total_passed += marks_with_grace

print(f"\nTotal marks of all passed students: {total_passed}")
```
